```python
path = "C:/Users/rohan/Jupyter Projects/MSEF/Clustering Workspace/"
infile = "Out_reformated_termsAndAllIds4clustering_inputForStep120211218.txt"
#file format: synonyms	keyword	freq	weight	snos

def fileToDictionary(path):
    weightDict = {}
    F = open(path, 'r')
    for line in F.readlines():
        row = line.split(",")
        if row[0] != "lineno": # exclude first line
            # print(row)
            weightDict.update({str(row[3]):float(row[11])})
    F.close()
    return weightDict
```

```python
wordWeightsFile = "wordWeights4Clustering.txt"
wordDict = fileToDictionary(path+wordWeightsFile)
```
